```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer  = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

tokenizer
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model = model,
    tokenizer = tokenizer,
    return_full_text=False, # if true, model will give input with output
    max_new_tokens=500, # some models might keep generating till they hit their context window limit
    do_sample=False  # weather model will choose sampling to choose next token.
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = "Create a funny joke about chickens."
output = generator(prompt, max_length=50)

print(output[0]["generated_text"]) 
```
