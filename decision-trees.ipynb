## Decision trees- these use tree data structure to

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
```

```python
df = pd.DataFrame(X)
df
```

```python
tree_classi = DecisionTreeClassifier(max_depth=2)

tree_classi.fit(X,y)
```

## Visualising tree

```python
from sklearn.tree import export_graphviz, plot_tree

plot_tree(tree_classi)
```

# Regression Trees

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X, y)
```

```python
plot_tree(tree_reg)
```
